````python
#Calcula el total de ventas por categoría de producto.

#DEFINIMOS LAS VARIABLES ANTES DE PASARLAS A NUMPY

categorias = df["Product Category"].to_numpy()
totales = df["Total Amount"].to_numpy()

````

````python
#Necesitamos establecer las categorias de forma unica para mejor respuesta.
categorias_uniques = np.unique(categorias)
````

````python
#Declaramos un diccionario para poder leer de mejor manera la informacion recolectada
ventas_c_p = {}

for categoria in categorias_uniques:
    filtrar = (categorias == categoria)
    total_sumado=  np.sum(totales[filtrar])
    ventas_c_p[categoria] = total_sumado

#Obtener el promedio desde las fechas
fech =df["Date"].to_numpy()

promedio_diario={}


#OBTENEMOS EL MAXIMO Y MINIMO DE LAS VENTAS
nombres_categorias = np.array(list(ventas_c_p.keys()))
valores_totales = np.array(list(ventas_c_p.values()))

ind_mayor= np.argmax(valores_totales)
ind_menor= np.argmin(valores_totales)

cat_top= nombres_categorias[ind_mayor]
valor_top = valores_totales[ind_mayor]

cat_baja = nombres_categorias[ind_menor]
valor_baja= valores_totales[ind_menor]



for cat, total in ventas_c_p.items():
    nfiltro = (categorias == cat)
    dias_c_ventas= np.unique(fech[nfiltro]).size
    #Calculando el promedio de ventas por dia segun sus cantidad
    if dias_c_ventas >0:
        prome_diario= total / dias_c_ventas
    else:
        prome_diario = 0
    promedio_diario[cat] = prome_diario
    
    
        
print("*"*35)
print("    Total ventas por categoria ")
print("*"*35)
for cat in categorias_uniques:
    total = ventas_c_p[cat]
    promedio = promedio_diario[cat]
    print(f"{cat}: ${total:,.2f} | Promedio diario: ${promedio:7,.2f}")
print("*"*35)


print("+"*35)
print("   Rendimientos")
print("+"*35)
print(f"Mayor valumen de ventas: {cat_top} (${valor_top:,.2f})")
print(f"Menor volumen de ventas: {cat_baja} (${valor_baja:,.2f})")
print("+"*35)
````

````python
def analisis_operacion_de_categoria(categoria_elegida, arr_categorias, arr_total, arra_cantidad):
    
    filtro = (arr_categorias == categoria_elegida)
    total_filtrado= arr_total[filtro]
    cantidades_filtrada= arra_cantidad[filtro]
    
    if len(total_filtrado) == 0:
        print(f"No se encontro ventas de esa categoria {categoria_elegida}")
        return
    
    imp= total_filtrado * 0.19
    ing_neto= total_filtrado - imp
    total_con_imp= total_filtrado + 5.00
    prec_unitario= total_filtrado / cantidades_filtrada
    
    print("&"*40)
    print(f" Analisis de la categoria: Categoria '{categoria_elegida}'")
    print(f" Se encontraron {len(total_filtrado)} transacciones.")
    print("&"*40)
    
    for i in range(3):
        if i < len(total_filtrado):
            print(f"Transaccion {i+1}")
            print(f" Total original: ${total_filtrado[i]:.2f}")
            print(f" Impuestos (19%): ${imp[i]:.2f}")
            print(f" Ingreso neto: ${ing_neto[i]:.2f}")
            print(f" Total + Envio: ${total_con_imp[i]:.2f}")
            print(f" Precio por unidad: ${prec_unitario[i]:.2f}")
            
````
